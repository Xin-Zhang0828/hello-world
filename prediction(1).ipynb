```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
test1=pd.read_csv('tcd ml 2019-20 income prediction training (with labels).csv')
test2=pd.read_csv('tcd ml 2019-20 income prediction test (without labels).csv')
test1.head(10)
```

```python
test1.columns=test2.columns
```

```python
test1.shape
```

```python
test2.shape
```

```python
test=pd.concat((test1,test2),axis=0)
test.shape
```

```python
test.head(10)
```

```python
test=test.drop(['Hair Color','Wears Glasses'],axis=1)

test.fillna(value={'Age':test.Age.median()}, inplace=True)
test.fillna(value={'Year of Record':test['Year of Record'].median()}, inplace=True)
test['Gender']=test['Gender'].replace('0','null')
test['University Degree']=test['University Degree'].replace('0','null')

test['Gender']=test['Gender'].fillna(method='bfill')
test['University Degree']=test['University Degree'].fillna(method='bfill')
test['Profession']=test['Profession'].fillna(method='backfill')
test.shape
```

```python
test.head(10)
```

```python
data_dummies = pd.get_dummies(test)
data_dummies.shape
```

```python
data_dummies.describe()
```

```python
test_1=data_dummies.iloc[0:111993,::]
test2=data_dummies.iloc[111993::,::]
```

```python
test_1.shape
```

```python
test2.shape
```

```python
train_missing = (test_1.isnull().sum()/len(test_1))*100
train_missing = train_missing.drop(train_missing[train_missing==0].index).sort_values(ascending=False)
miss_data = pd.DataFrame({'缺失百分比':train_missing})
miss_data
```

```python
test_1.dropna(inplace=True)
```

```python
y = test_1['Income']
X = test_1.drop(['Income','Gender_null','University Degree_null'],axis=1)
print('data shape: {0}; no. positive: {1}; no. negative: {2}'.format(
    X.shape, y[y==1].shape[0], y[y==0].shape[0]))
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#from sklearn.ensemble import RandomForestRegressor
#model =RandomForestRegressor(n_estimators=10)
#model.fit(X_train, y_train)
#train_score = model.score(X_train, y_train)
#test_score = model.score(X_test, y_test)
#print('train score: {train_score:.6f}; test score: {test_score:.6f}'.format(
#     train_score=train_score, test_score=test_score))
```

```python
# Linear Regression
from sklearn import linear_model
model =linear_model.LinearRegression()
model.fit(X_train, y_train)
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)
print('train score: {train_score:.6f}; test score: {test_score:.6f}'.format(
   train_score=train_score, test_score=test_score))
```
